```python
import pandas as pd
from sqlalchemy import create_engine


df = pd.read_csv('sales_pipeline.csv')

# 2. Create the connection (Replace with your Postgres credentials)
# Format: postgresql://username:password@host:port/database_name
engine = create_engine('postgresql://postgres:jeremycummins@localhost:5433/jeremycummins')

# 3. Push to Postgres
# 'if_exists=replace' creates the table for you automatically
df.to_sql('sales_pipeline', engine, if_exists='replace', index=False)

print("Success! Data ingested into PostgreSQL.")
```
