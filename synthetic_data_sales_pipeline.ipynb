```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

def generate_pipeline_data(n_deals=200):
    np.random.seed(42) # For reproducibility
    
    stages = {
        '1-Discovery': 0.10,
        '2-Qualification': 0.25,
        '3-Proposal': 0.50,
        '4-Negotiation': 0.75,
        '5-Closed Won': 1.00
    }
    
    segments = ['Enterprise', 'Mid-Market', 'SMB']
    data = []
    
    for i in range(n_deals):
        stage = np.random.choice(list(stages.keys()))
        prob = stages[stage]
        
        # Larger deals for Enterprise, smaller for SMB
        segment = np.random.choice(segments, p=[0.2, 0.5, 0.3])
        if segment == 'Enterprise':
            value = np.random.uniform(100000, 500000)
        elif segment == 'Mid-Market':
            value = np.random.uniform(25000, 100000)
        else:
            value = np.random.uniform(5000, 25000)
            
        # Random close dates over the next 6 months
        close_date = datetime(2024, 1, 1) + timedelta(days=np.random.randint(0, 180))
        
        data.append([
            f"OPP_{i:04d}", 
            f"ACC_{np.random.randint(0, 50)}", # Linking to existing accounts
            stage, 
            round(value, 2), 
            prob, 
            close_date.date(),
            segment
        ])
        
    return pd.DataFrame(data, columns=[
        'opportunity_id', 'account_id', 'deal_stage', 
        'forecasted_value', 'probability', 'close_date', 'segment'
    ])

pipeline_df = generate_pipeline_data()
pipeline_df.to_csv('sales_pipeline.csv', index=False)
print(f"Generated {len(pipeline_df)} sales opportunities.")
```
